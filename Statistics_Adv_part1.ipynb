##Theory:-                                                                

1. What is a random variable in probability theory?
--> A random variable is a variable whose possible values are determined by the outcome of a random experiment.
    It can be either discrete (taking specific values) or continuous (taking any value within a range).

2. What are the types of random variables?
--> There are two main types:

 a) Discrete random variables: Can take specific, distinct values (e.g., number of heads in coin flips).
 b) Continuous random variables: Can take any value within a given range (e.g., height, weight).

3. What is the difference between discrete and continuous distributions?
--> Discrete distribution: Deals with discrete random variables and is characterized by a probability mass function (PMF).
    Continuous distribution: Deals with continuous random variables and is characterized by a probability density function (PDF).

4. What are probability distribution functions (PDF)?
--> A probability distribution function gives the probabilities of all possible outcomes for a random variable. 
    For discrete variables, it is the probability mass function (PMF); for continuous variables, it is the probability density function (PDF).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
-->  The PDF gives the probability of individual outcomes, while the CDF gives the cumulative probability of a variable being less than or equal to a certain value.

6. What is a discrete uniform distribution?
--> A discrete uniform distribution is a distribution where all outcomes have the same probability. It is used for scenarios like rolling a fair die.

7. What are the key properties of a Bernoulli distribution?
--> A Bernoulli distribution models a binary outcome, typically representing a success (1) or failure (0) with 
    probabilities ð‘ and 1âˆ’p respectively.

8. What is the binomial distribution, and how is it used in probability?
--> A binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. 
    It is used for binary events with a fixed probability.

9. What is the Poisson distribution and where is it applied?
--> The Poisson distribution models the number of events occurring within a fixed interval of time or space. 
    It is often applied in fields like queuing theory and natural events (e.g., phone calls to a center).

10. What is a continuous uniform distribution?
--> A continuous uniform distribution is a probability distribution where every value in a continuous range has an equal probability of occurring.

11. What are the characteristics of a normal distribution?
--> A normal distribution is symmetric, bell-shaped, with its peak at the mean. It is fully characterized by its mean (Âµ) and standard deviation (Ïƒ).

12. What is the standard normal distribution, and why is it important?
--> The standard normal distribution is a normal distribution with a mean of 0 and a standard deviation of 1. 
    It is important because it allows comparisons across different normal distributions through Z-scores.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
--> The CLT states that the sampling distribution of the sample mean will be approximately normal, regardless of the population distribution, 
    as the sample size increases. It is fundamental for hypothesis testing and confidence intervals.

14. How does the Central Limit Theorem relate to the normal distribution?
--> The CLT shows that, with a sufficiently large sample size, the distribution of sample means approaches a normal distribution, 
    even if the original data is not normally distributed.

15. What is the application of Z statistics in hypothesis testing?
--> Z-statistics are used to determine how many standard deviations a sample mean is from the population mean. 
    It is used in hypothesis testing to evaluate the statistical significance of a sample.

16. How do you calculate a Z-score, and what does it represent?
--> A Z-score is calculated as Z= Xâˆ’Î¼/ðœŽ, where X is the data point, Î¼ is the mean, and Ïƒ is the standard deviation. 
    It represents how many standard deviations a data point is from the mean.

17. What are point estimates and interval estimates in statistics?

--> Point estimate: A single value estimate of a population parameter (e.g., sample mean).
    Interval estimate: A range of values within which the population parameter is likely to lie (e.g., confidence interval).

18. What is the significance of confidence intervals in statistical analysis?
--> Confidence intervals provide a range of values for a population parameter, indicating the level of uncertainty or precision in the estimate.

19. What is the relationship between a Z-score and a confidence interval?
--> The Z-score is used to determine the critical value for the confidence interval, defining the margin of error.

20. How are Z-scores used to compare different distributions?
--> Z-scores standardize data, allowing comparisons of values from different distributions by converting them to a common scale.

21. What are the assumptions for applying the Central Limit Theorem?
--> The data should be independent, and the sample size should be large enough (typically n > 30). The population distribution should not 
    be heavily skewed.

22. What is the concept of expected value in a probability distribution?
--> The expected value is the long-term average or mean of a random variable, representing the "center" of its probability distribution.

23. How does a probability distribution relate to the expected outcome of a random variable?
--> A probability distribution describes the likelihood of different outcomes, and the expected value represents the mean of those outcomes
    weighted by their probabilities.
